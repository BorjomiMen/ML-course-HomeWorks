## Общее представление о кластеризации

Кластеризация — это метод обучения без учителя, целью которого является разделение набора данных на группы (кластеры) таким образом, чтобы объекты внутри одного кластера были максимально похожи друг на друга, а объекты из разных кластеров — максимально отличались. Эти методы широко применяются в различных областях для выявления скрытых структур в данных.

## Метод KMeans

### Принцип работы
KMeans — центроидный алгоритм, который делит данные на K кластеров, минимизируя сумму квадратов расстояний от точек до центроидов кластеров.

### Преимущества
- Высокая скорость работы на больших объемах данных
- Простота реализации и интерпретации
- Масштабируемость на большие наборы данных
- Четкие, выпуклые кластеры

### Недостатки
- Требует заранее заданного количества кластеров (K)
- Чувствителен к выбросам
- Предполагает сферическую форму кластеров
- Зависит от начальной инициализации центроидов

### Области применения
- Сегментация клиентов в маркетинге
- Анализ изображений (сжатие цветов)
- Документная кластеризация
- Биоинформатика (кластеризация генов)

## Метод DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

### Принцип работы
DBSCAN группирует точки на основе плотности, образуя кластеры как области с высокой плотностью точек, разделенные областями с низкой плотностью.

### Преимущества
- Не требует заранее заданного числа кластеров
- Устойчив к выбросам (автоматически их идентифицирует)
- Способен находить кластеры произвольной формы
- Не зависит от начальной инициализации

### Недостатки
- Проблемы с кластерами разной плотности
- Чувствителен к выбору параметров (eps и min_samples)
- Сложности с данными высокой размерности
- Медленнее KMeans на больших объемах данных

### Области применения
- Обнаружение аномалий и выбросов
- Анализ географических данных
- Сегментация изображений
- Анализ сетевого трафика для обнаружения вторжений
- Социальный анализ (выявление сообществ)

## Метод Agglomerative Clustering (Иерархическая агломеративная кластеризация)

### Принцип работы
Иерархический метод, который строит дерево кластеров (дендрограмму), последовательно объединяя наиболее похожие кластеры.

### Преимущества
- Не требует заранее заданного числа кластеров
- Визуализация в виде дендрограммы
- Устойчивость к шуму
- Возможность выбора любого уровня иерархии

### Недостатки
- Вычислительная сложность O(n³) или O(n² log n)
- Чувствительность к выбору меры расстояния и метода связи
- Необратимость слияний
- Проблемы с масштабированием на большие наборы данных

### Области применения
- Биология (филогенетические деревья)
- Лингвистика (кластеризация языков)
- Социальные сети (анализ сообществ)
- Маркетинг (сегментация клиентов с иерархией)
- Дендрограммы в любых областях, где важна иерархия

## Ключевые различия между методами

### 1. Подход к формированию кластеров
- **KMeans**: Центроидный, основан на минимизации расстояний до центров
- **DBSCAN**: Плотностный, основан на группировке близко расположенных точек
- **Agglomerative**: Иерархический, основан на последовательном слиянии

### 2. Форма кластеров
- **KMeans**: Сферическая, выпуклая
- **DBSCAN**: Произвольная форма
- **Agglomerative**: Различные формы в зависимости от меры расстояния

### 3. Работа с выбросами
- **KMeans**: Чувствителен, выбросы могут исказить центроиды
- **DBSCAN**: Автоматически идентифицирует как шум
- **Agglomerative**: Относительно устойчив

### 4. Требование количества кластеров
- **KMeans**: Требует заранее заданного K
- **DBSCAN**: Не требует
- **Agglomerative**: Не требует (можно выбрать на любом уровне)

### 5. Вычислительная сложность
- **KMeans**: O(n*k*i) — самый быстрый
- **DBSCAN**: O(n log n) при использовании индексов
- **Agglomerative**: O(n³) или O(n² log n) — самый медленный

## Практические рекомендации по выбору метода

### Выбирайте KMeans, когда:
- Данные имеют сферическую форму кластеров
- Известно или можно оценить количество кластеров
- Требуется быстрая обработка больших объемов данных
- Нужны четкие, легко интерпретируемые результаты

### Выбирайте DBSCAN, когда:
- Форма кластеров произвольная или неизвестна
- Имеются выбросы, которые нужно исключить
- Неизвестно количество кластеров
- Важна устойчивость к шуму

### Выбирайте Agglomerative Clustering, когда:
- Важна иерархическая структура данных
- Нужна визуализация в виде дендрограммы
- Размер данных умеренный
- Требуется анализ на разных уровнях детализации

## Комбинированный подход в практических задачах

В реальных проектах часто используется комбинация методов:
1. DBSCAN для предварительной очистки от выбросов
2. KMeans для основной кластеризации очищенных данных
3. Agglomerative для углубленного анализа структуры

Такой подход позволяет использовать сильные стороны каждого метода и получать более надежные результаты.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

print("=" * 80)
print("ШАГ 1: ЗАГРУЗКА И ПРЕДВАРИТЕЛЬНЫЙ АНАЛИЗ")
print("=" * 80)

file_path = "Video_Games_Sales_as_at_22_Dec_2016.csv"

try:
    import kagglehub
    from kagglehub import KaggleDatasetAdapter
    df = kagglehub.load_dataset(
        KaggleDatasetAdapter.PANDAS,
        "rush4ratio/video-game-sales-with-ratings",
        file_path,
    )
except ImportError:
    print("Установка kagglehub...")
    import subprocess, sys
    subprocess.check_call([sys.executable, "-m", "pip", "install", "kagglehub"])
    import kagglehub
    from kagglehub import KaggleDatasetAdapter
    df = kagglehub.load_dataset(
        KaggleDatasetAdapter.PANDAS,
        "rush4ratio/video-game-sales-with-ratings",
        file_path,
    )

print(f"Размер датасета: {df.shape}")
print(f"Количество признаков: {df.shape[1]}")
print("\nПервые 5 строк:")
print(df.head())

print("\n" + "=" * 80)
print("ШАГ 2: КОМБИНИРОВАННАЯ ПРЕДОБРАБОТКА ДАННЫХ")
print("=" * 80)

df_clean = df.copy()

df_clean['User_Score'] = pd.to_numeric(df_clean['User_Score'], errors='coerce')


print("\nЗаполнение пропущенных значений...")

# Числовые признаки - медиана
num_cols_to_fill = ['Year_of_Release', 'Critic_Score', 'Critic_Count', 
                    'User_Score', 'User_Count', 'NA_Sales', 'EU_Sales', 
                    'JP_Sales', 'Other_Sales', 'Global_Sales']

for col in num_cols_to_fill:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna(df_clean[col].median())

cat_cols_to_fill = ['Publisher', 'Developer', 'Rating']
for col in cat_cols_to_fill:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna('Unknown')

initial_size = len(df_clean)
df_clean = df_clean.dropna(subset=['Genre'])
print(f"\nУдалено строк с пропущенным жанром: {initial_size - len(df_clean)}")

print("\nОстаточные пропущенные значения:")
print(df_clean.isnull().sum()[df_clean.isnull().sum() > 0])

print("\n" + "=" * 80)
print("ШАГ 3: СОЗДАНИЕ НОВЫХ ПРИЗНАКОВ ДЛЯ КЛАСТЕРИЗАЦИИ")
print("=" * 80)

df_clean['Overall_Rating'] = (df_clean['Critic_Score'] * 0.7 + df_clean['User_Score'] * 10 * 0.3)
df_clean['Sales_Per_Year'] = df_clean['Global_Sales'] / (2020 - df_clean['Year_of_Release']).clip(1)

df_clean['NA_Sales_Ratio'] = df_clean['NA_Sales'] / df_clean['Global_Sales'].replace(0, 0.001)
df_clean['EU_Sales_Ratio'] = df_clean['EU_Sales'] / df_clean['Global_Sales'].replace(0, 0.001)
df_clean['JP_Sales_Ratio'] = df_clean['JP_Sales'] / df_clean['Global_Sales'].replace(0, 0.001)

df_clean = df_clean[df_clean['Global_Sales'] > 0.01]
df_clean = df_clean.dropna(subset=['Overall_Rating', 'Sales_Per_Year'])

print(f"Данных после очистки: {df_clean.shape}")
print(f"Диапазон продаж: {df_clean['Global_Sales'].min():.2f} - {df_clean['Global_Sales'].max():.2f} млн.")
print(f"Диапазон оценок: {df_clean['Overall_Rating'].min():.1f} - {df_clean['Overall_Rating'].max():.1f}")

print("\n" + "=" * 80)
print("ШАГ 4: ПОДГОТОВКА ПРИЗНАКОВ ДЛЯ КЛАСТЕРИЗАЦИИ")
print("=" * 80)

features = [
    'Global_Sales', 'NA_Sales_Ratio', 'EU_Sales_Ratio', 'JP_Sales_Ratio',
    'Overall_Rating', 'Critic_Count', 'User_Count', 'Sales_Per_Year', 'Year_of_Release'
]

print("Используемые признаки для кластеризации:")
for i, feat in enumerate(features, 1):
    print(f"{i:2}. {feat}")

X = df_clean[features].copy()

X['Global_Sales'] = np.log1p(X['Global_Sales'])
X['Sales_Per_Year'] = np.log1p(X['Sales_Per_Year'])

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

print(f"\nРазмерность данных для кластеризации: {X_scaled.shape}")

print("\n" + "=" * 80)
print("ШАГ 5: ОПРЕДЕЛЕНИЕ ОПТИМАЛЬНОГО ЧИСЛА КЛАСТЕРОВ")
print("=" * 80)

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia = []
silhouette_scores = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=20)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    silhouette_avg = silhouette_score(X_scaled, kmeans.labels_)
    silhouette_scores.append(silhouette_avg)


fig1, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
ax1.plot(K_range, inertia, 'bo-')
ax1.set_xlabel('Количество кластеров')
ax1.set_ylabel('Инерция (Inertia)')
ax1.set_title('Метод локтя для KMeans')
ax1.grid(True, alpha=0.3)

ax2.plot(K_range, silhouette_scores, 'ro-')
ax2.set_xlabel('Количество кластеров')
ax2.set_ylabel('Средний коэффициент силуэта')
ax2.set_title('Метод силуэта для KMeans')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('optimal_clusters_fixed.png', dpi=100, bbox_inches='tight')
plt.show()

optimal_k_silhouette = K_range[np.argmax(silhouette_scores)]
print(f"Оптимальное количество кластеров по силуэту: {optimal_k_silhouette}")
optimal_k = optimal_k_silhouette

print("\n" + "=" * 80)
print("ШАГ 6: КЛАСТЕРИЗАЦИЯ РАЗНЫМИ МЕТОДАМИ")
print("=" * 80)

from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.neighbors import NearestNeighbors
import scipy.cluster.hierarchy as sch


pca_2d = PCA(n_components=2, random_state=42)
X_pca_2d = pca_2d.fit_transform(X_scaled)

pca_3d = PCA(n_components=3, random_state=42)
X_pca_3d = pca_3d.fit_transform(X_scaled)

pca_cluster = PCA(n_components=0.85, random_state=42)
X_pca_cluster = pca_cluster.fit_transform(X_scaled)
print(f"Для кластеризации используется {X_pca_cluster.shape[1]} компонент PCA")

print("\n1. KMeans кластеризация...")
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=20)
kmeans_labels = kmeans.fit_predict(X_scaled)
df_clean['KMeans_Cluster'] = kmeans_labels

print("2. DBSCAN кластеризация...")

def find_optimal_eps(data, min_samples, percentile=90):
    """Находит оптимальное значение eps для DBSCAN"""
    neighbors = NearestNeighbors(n_neighbors=min_samples)
    neighbors_fit = neighbors.fit(data)
    distances, indices = neighbors_fit.kneighbors(data)
    k_distances = np.sort(distances[:, min_samples-1])
    eps_value = np.percentile(k_distances, percentile)
    return eps_value, k_distances


min_samples_dbscan = 10
optimal_eps, k_distances = find_optimal_eps(X_pca_cluster, min_samples_dbscan, percentile=85)
print(f"   Автоматически подобранный eps: {optimal_eps:.3f}")

plt.figure(figsize=(10, 6))
plt.plot(k_distances)
plt.axhline(y=optimal_eps, color='r', linestyle='--', label=f'eps = {optimal_eps:.3f}')
plt.xlabel('Точки данных (отсортированные)')
plt.ylabel(f'Расстояние до {min_samples_dbscan}-го соседа')
plt.title('K-distance график для DBSCAN')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('dbscan_kdistance_fixed.png', dpi=100, bbox_inches='tight')
plt.show()

dbscan = DBSCAN(eps=optimal_eps, min_samples=min_samples_dbscan, metric='euclidean')
dbscan_labels = dbscan.fit_predict(X_pca_cluster)
df_clean['DBSCAN_Cluster'] = dbscan_labels

print("3. Agglomerative Clustering...")
agglomerative = AgglomerativeClustering(n_clusters=optimal_k, linkage='ward')
agglomerative_labels = agglomerative.fit_predict(X_scaled)
df_clean['Agglomerative_Cluster'] = agglomerative_labels

print("   Создание дендрограммы...")
plt.figure(figsize=(12, 6))
sample_size = min(100, len(X_scaled))
sample_indices = np.random.choice(len(X_scaled), sample_size, replace=False)
X_sample = X_scaled.iloc[sample_indices].values

dendrogram = sch.dendrogram(sch.linkage(X_sample, method='ward'),
                            truncate_mode='level', p=5)
plt.title(f'Дендрограмма Agglomerative Clustering (выборка из {sample_size} точек)')
plt.xlabel('Точки данных')
plt.ylabel('Евклидово расстояние')
plt.axhline(y=7, color='r', linestyle='--', label=f'Разрез для {optimal_k} кластеров')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('dendrogram_fixed.png', dpi=100, bbox_inches='tight')
plt.show()

print(f"\nСТАТИСТИКА КЛАСТЕРОВ:")
print("-" * 60)
print(f"KMeans:           {len(np.unique(kmeans_labels))} кластеров")
dbscan_unique = np.unique(dbscan_labels)
dbscan_clusters = len(dbscan_unique) - (1 if -1 in dbscan_unique else 0)
print(f"DBSCAN:           {dbscan_clusters} кластеров ({np.sum(dbscan_labels == -1)} выбросов, {np.sum(dbscan_labels == -1)/len(dbscan_labels)*100:.1f}%)")
print(f"Agglomerative:    {len(np.unique(agglomerative_labels))} кластеров")

print("\n" + "=" * 80)
print("ШАГ 7: ВИЗУАЛИЗАЦИЯ РЕЗУЛЬТАТОВ КЛАСТЕРИЗАЦИИ")
print("=" * 80)

from mpl_toolkits.mplot3d import Axes3D

fig1 = plt.figure(figsize=(18, 8))

methods = [
    ('KMeans', kmeans_labels, 'viridis'),
    ('DBSCAN', dbscan_labels, 'plasma'), 
    ('Agglomerative', agglomerative_labels, 'turbo')
]


for idx, (name, labels, cmap) in enumerate(methods, 1):
    ax = fig1.add_subplot(2, 3, idx)
    
    if name == 'DBSCAN':
        outlier_mask = labels == -1
        cluster_mask = ~outlier_mask
        
        if np.any(outlier_mask):
            ax.scatter(X_pca_2d[outlier_mask, 0], X_pca_2d[outlier_mask, 1], 
                      c='gray', s=10, alpha=0.3, label='Выбросы', edgecolor='k', linewidth=0.2)
        
        if np.any(cluster_mask):
            scatter = ax.scatter(X_pca_2d[cluster_mask, 0], X_pca_2d[cluster_mask, 1], 
                                c=labels[cluster_mask], cmap=cmap, s=15, alpha=0.6, 
                                edgecolor='k', linewidth=0.2)
            plt.colorbar(scatter, ax=ax, label='Cluster')
    else:
        scatter = ax.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=labels,
                            cmap=cmap, s=15, alpha=0.6, edgecolor='k', linewidth=0.2)
        plt.colorbar(scatter, ax=ax, label='Cluster')
    
    ax.set_xlabel('PCA Component 1')
    ax.set_ylabel('PCA Component 2')
    ax.set_title(f'{name} Clustering (2D PCA)')

for idx, (name, labels, cmap) in enumerate(methods, 4):
    ax = fig1.add_subplot(2, 3, idx, projection='3d')
    
    if name == 'DBSCAN':
        outlier_mask = labels == -1
        cluster_mask = ~outlier_mask
        
        if np.any(outlier_mask):
            ax.scatter(X_pca_3d[outlier_mask, 0], X_pca_3d[outlier_mask, 1], X_pca_3d[outlier_mask, 2],
                      c='gray', s=5, alpha=0.2, label='Выбросы', depthshade=True)
        
        if np.any(cluster_mask):
            scatter = ax.scatter(X_pca_3d[cluster_mask, 0], X_pca_3d[cluster_mask, 1], X_pca_3d[cluster_mask, 2],
                                c=labels[cluster_mask], cmap=cmap, s=8, alpha=0.6, depthshade=True)
    else:
        scatter = ax.scatter(X_pca_3d[:, 0], X_pca_3d[:, 1], X_pca_3d[:, 2],
                            c=labels, cmap=cmap, s=8, alpha=0.6, depthshade=True)
    
    ax.set_xlabel('PCA 1')
    ax.set_ylabel('PCA 2')
    ax.set_zlabel('PCA 3')
    ax.set_title(f'{name} Clustering (3D PCA)')

plt.tight_layout()
plt.savefig('clustering_comparison_fixed.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n" + "=" * 80)
print("ШАГ 8: АНАЛИЗ И ИНТЕРПРЕТАЦИЯ КЛАСТЕРОВ")
print("=" * 80)

from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

print("\nСРАВНЕНИЕ МЕТОДОВ ПО МЕТРИКАМ:")
print("-" * 60)

metrics_data = []
methods_info = [
    ('KMeans', kmeans_labels, X_scaled),
    ('DBSCAN', dbscan_labels, X_pca_cluster),
    ('Agglomerative', agglomerative_labels, X_scaled)
]

for name, labels, data in methods_info:
    if name == 'DBSCAN':
        valid_mask = labels != -1
        n_clusters = len(np.unique(labels[valid_mask]))
        n_outliers = np.sum(labels == -1)
        
        if n_clusters > 1 and np.sum(valid_mask) > 1:
            silhouette = silhouette_score(data[valid_mask], labels[valid_mask])
            calinski = calinski_harabasz_score(data[valid_mask], labels[valid_mask])
            davies = davies_bouldin_score(data[valid_mask], labels[valid_mask])
        else:
            silhouette = calinski = davies = np.nan
    else:
        n_clusters = len(np.unique(labels))
        n_outliers = 0
        silhouette = silhouette_score(data, labels)
        calinski = calinski_harabasz_score(data, labels)
        davies = davies_bouldin_score(data, labels)
    
    metrics_data.append({
        'Method': name,
        'Clusters': n_clusters,
        'Outliers': n_outliers,
        'Silhouette': silhouette,
        'Calinski_Harabasz': calinski,
        'Davies_Bouldin': davies
    })

metrics_df = pd.DataFrame(metrics_data)
print(metrics_df.to_string(index=False, float_format=lambda x: f'{x:.3f}' if not pd.isna(x) else 'N/A'))

print(f"\nАНАЛИЗ KMEANS КЛАСТЕРОВ ({optimal_k} кластеров):")
print("-" * 60)

X_original = pd.DataFrame(scaler.inverse_transform(X_scaled), columns=X.columns)
X_original['Global_Sales'] = np.expm1(X_original['Global_Sales'])
X_original['Sales_Per_Year'] = np.expm1(X_original['Sales_Per_Year'])

for cluster_id in range(optimal_k):
    cluster_mask = kmeans_labels == cluster_id
    
    if np.sum(cluster_mask) > 0:
        cluster_data = df_clean[cluster_mask]
        
        print(f"\nКластер {cluster_id}:")
        print(f"  Размер: {np.sum(cluster_mask):,} игр ({np.sum(cluster_mask)/len(kmeans_labels)*100:.1f}%)")
        print(f"  Средние продажи: {X_original.loc[cluster_mask, 'Global_Sales'].mean():.2f} млн")
        print(f"  Средний рейтинг: {X_original.loc[cluster_mask, 'Overall_Rating'].mean():.1f}")
        print(f"  Средний год выпуска: {df_clean.loc[cluster_mask, 'Year_of_Release'].mean():.0f}")
        

        na_ratio = X_original.loc[cluster_mask, 'NA_Sales_Ratio'].mean()
        eu_ratio = X_original.loc[cluster_mask, 'EU_Sales_Ratio'].mean()
        jp_ratio = X_original.loc[cluster_mask, 'JP_Sales_Ratio'].mean()
        print(f"  Региональные соотношения: NA={na_ratio:.1%}, EU={eu_ratio:.1%}, JP={jp_ratio:.1%}")
        
        
        top_genres = cluster_data['Genre'].value_counts().head(3)
        print(f"  Топ жанры: {', '.join([f'{g} ({c})' for g, c in top_genres.items()])}")
        
        
        top_publishers = cluster_data['Publisher'].value_counts().head(3)
        print(f"  Топ издатели: {', '.join([f'{p} ({c})' for p, c in top_publishers.items()])}")


print("\n" + "=" * 80)
print("ШАГ 9: СРАВНЕНИЕ КЛАСТЕРОВ С ИСТИННЫМИ МЕТКАМИ (ЖАНРАМИ)")
print("=" * 80)

from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score, homogeneity_score, completeness_score
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()
true_labels = le.fit_transform(df_clean['Genre'])

print(f"Количество уникальных жанров: {len(le.classes_)}")
print(f"Список жанров: {list(le.classes_)}")


print("\nСРАВНЕНИЕ КЛАСТЕРОВ С ИСТИННЫМИ ЖАНРАМИ:")
print("-" * 70)

comparison_results = []

methods_comparison = [
    ('KMeans', kmeans_labels),
    ('Agglomerative', agglomerative_labels),
]

for method_name, cluster_labels in methods_comparison:
    
    ari = adjusted_rand_score(true_labels, cluster_labels)
    nmi = normalized_mutual_info_score(true_labels, cluster_labels)
    homogeneity = homogeneity_score(true_labels, cluster_labels)
    completeness = completeness_score(true_labels, cluster_labels)
    
    comparison_results.append({
        'Method': method_name,
        'Adjusted_Rand_Index': ari,
        'Normalized_MI': nmi,
        'Homogeneity': homogeneity,
        'Completeness': completeness
    })
    
    print(f"\n{method_name}:")
    print(f"  • Adjusted Rand Index: {ari:.4f}")
    print(f"  • Normalized Mutual Information: {nmi:.4f}")
    print(f"  • Homogeneity: {homogeneity:.4f}")
    print(f"  • Completeness: {completeness:.4f}")

print(f"\nDBSCAN (с учетом выбросов как отдельного кластера):")
dbscan_ari = adjusted_rand_score(true_labels, dbscan_labels)
dbscan_nmi = normalized_mutual_info_score(true_labels, dbscan_labels)
print(f"  • Adjusted Rand Index: {dbscan_ari:.4f}")
print(f"  • Normalized Mutual Information: {dbscan_nmi:.4f}")

comparison_df = pd.DataFrame(comparison_results)
print("\nСВОДНАЯ ТАБЛИЦА СРАВНЕНИЯ С ИСТИННЫМИ МЕТКАМИ:")
print("-" * 70)
print(comparison_df.to_string(index=False, float_format=lambda x: f'{x:.4f}'))

print("\n" + "=" * 80)
print("ШАГ 10: СРАВНЕНИЕ KMEANS И AGGLOMERATIVE CLUSTERING")
print("=" * 80)

from scipy.stats import mode

print("\nСОПОСТАВЛЕНИЕ КЛАСТЕРОВ KMEANS И AGGLOMERATIVE:")
print("-" * 60)

contingency_table = pd.crosstab(kmeans_labels, agglomerative_labels)
print("\nМатрица сопряженности кластеров (KMeans строки, Agglomerative столбцы):")
print(contingency_table)

max_overlap = 0
print("\nСООТВЕТСТВИЕ КЛАСТЕРОВ ПО МЕТОДАМ:")
print("-" * 60)

for kmeans_cluster in range(optimal_k):
    if kmeans_cluster in contingency_table.index:
        max_in_row = contingency_table.loc[kmeans_cluster].max()
        total_in_row = contingency_table.loc[kmeans_cluster].sum()
        overlap_percent = max_in_row / total_in_row * 100 if total_in_row > 0 else 0
        max_overlap += max_in_row
        
        corresponding_agg_cluster = contingency_table.loc[kmeans_cluster].idxmax()
        print(f"KMeans кластер {kmeans_cluster}: {overlap_percent:.1f}% совпадает с Agglomerative кластер {corresponding_agg_cluster}")

total_overlap_percent = max_overlap / len(kmeans_labels) * 100
print(f"\nОбщее совпадение кластеров между KMeans и Agglomerative: {total_overlap_percent:.1f}%")

print("\n" + "=" * 80)
print("ШАГ 11: АНАЛИЗ РАСПРЕДЕЛЕНИЯ ЖАНРОВ В КЛАСТЕРАХ")
print("=" * 80)

print("\nРАСПРЕДЕЛЕНИЕ ЖАНРОВ ПО КЛАСТЕРАМ KMEANS:")
print("-" * 60)

for cluster_id in range(optimal_k):
    cluster_mask = kmeans_labels == cluster_id
    cluster_genres = df_clean.loc[cluster_mask, 'Genre']
    
    if len(cluster_genres) > 0:
        genre_counts = cluster_genres.value_counts()
        total_games = len(cluster_genres)
        
        print(f"\nКластер {cluster_id} ({total_games} игр):")
        print(f"Топ-5 жанров:")
        for genre, count in genre_counts.head(5).items():
            percentage = count / total_games * 100
            print(f"  {genre}: {count} игр ({percentage:.1f}%)")
```
